# Difference Between MOD03 and MOD14 MODIS Products

The main difference between MOD03 and MOD14 is the type of data they provide, as both are different MODIS (Moderate Resolution Imaging Spectroradiometer) products designed for different purposes.

## 1. MOD03: Geolocation Product
**Product Description:**  
MOD03 is the MODIS Level 1B Geolocation product. It provides essential geolocation information for each MODIS scan line, including latitude, longitude, height, and other geolocation data necessary to interpret and map MODIS observations.

**Purpose:**  
MOD03 data is used to map the measurements from other MODIS instruments onto the Earth’s surface. It provides detailed information that helps in the precise geolocation of MODIS pixels.

**Data Content:**
- Latitude and longitude
- Solar and satellite zenith and azimuth angles
- Sensor position and view angles
- Elevation of the Earth’s surface
- Time of acquisition

**Applications:**  
The MOD03 product is crucial for interpreting the geospatial location of any data collected by other MODIS products, such as sea surface temperature, land cover, or vegetation index data.

---

## 2. MOD14: Thermal Anomalies and Fire Detection Product
**Product Description:**  
MOD14 is the MODIS Fire and Thermal Anomalies product. It is designed to detect thermal anomalies, primarily fires, using infrared bands to identify areas with abnormally high surface temperatures.

**Purpose:**  
This product identifies active fire locations globally. It uses a combination of brightness temperature thresholds and contextual algorithms to detect potential fires and hotspots.

**Data Content:**
- Fire mask, which labels pixels as:
  - Fire
  - Water
  - Land
  - Cloud cover
- Fire radiative power (FRP) for detected fires
- Quality assessment information
- Confidence levels (low, medium, high) of detected fire pixels

**Applications:**  
MOD14 is widely used for fire detection, wildfire management, and thermal anomaly monitoring. It helps in detecting forest fires, urban fires, and other thermal events from space.

```python
from pyhdf.SD import SD, SDC
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Open the MOD14 file (active fire product)
fire_file = SD('MOD14.A2018312.1950.061.2021344091856.hdf', SDC.READ)

# Step 2: List all datasets in the MOD14 file (optional, for exploration)
fire_datasets = fire_file.datasets()
print(fire_datasets)

# Extract the fire mask or relevant fire dataset (example: 'FireMask')
fire_mask = fire_file.select('fire mask')[:]  # Replace 'FireMask' with the correct dataset name if different
print(f"Fire mask data shape: {fire_mask.shape}")

# Extract fire pixel latitude and longitude from the MOD14 file (fp_latitude and fp_longitude)
# Ensure the correct names for these datasets; here assuming 'FP_Latitude' and 'FP_Longitude'
fp_latitude = fire_file.select('FP_latitude')[:]  # Replace with correct name
fp_longitude = fire_file.select('FP_longitude')[:]  # Replace with correct name
print(f"FP Latitude shape: {fp_latitude.shape}")
print(f"FP Longitude shape: {fp_longitude.shape}")

# Step 3: Open the MOD03 file (geolocation product)
geo_file_path = 'MOD03.A2018312.1950.061.2018313011438.hdf'
geo_file = SD(geo_file_path, SDC.READ)

# Read the Latitude and Longitude datasets from the geolocation file
latitude_data = geo_file.select('Latitude')[:]
longitude_data = geo_file.select('Longitude')[:]

print(f"Latitude data shape: {latitude_data.shape}")
print(f"Longitude data shape: {longitude_data.shape}")

# Step 4: Filter fire pixels with fire mask values 7, 8, and 9
fire_pixels_mask = (fire_mask == 7) | (fire_mask == 8) | (fire_mask == 9)

# Extract the latitude and longitude of those fire pixels from the MOD03 geolocation data
fire_latitudes = latitude_data[fire_pixels_mask]
fire_longitudes = longitude_data[fire_pixels_mask]

# Step 5: Compare the geolocated fire pixels with the fp_latitude and fp_longitude from the MOD14 file
# Here, we'll calculate the difference between the fp_latitude/fp_longitude and the geolocated data
latitude_diff = np.abs(fp_latitude - fire_latitudes)
longitude_diff = np.abs(fp_longitude - fire_longitudes)

# Set tolerance for the comparison (small value to account for minor differences)
tolerance = 0.01  # Adjust this based on the required precision

# Check if the latitude and longitude values match within the tolerance
latitude_match = np.all(latitude_diff < tolerance)
longitude_match = np.all(longitude_diff < tolerance)

print(f"Latitude match within tolerance: {latitude_match}")
print(f"Longitude match within tolerance: {longitude_match}")

# Step 6: Plot the fire data (fire mask 7, 8, 9) on the corresponding latitude/longitude map
plt.figure(figsize=(10, 6))

# Create a scatter plot of the geolocated fire pixels
plt.scatter(fire_longitudes, fire_latitudes, color='red', label='Fire Pixels (Geolocated)', s=1)

# Create a scatter plot of the fire pixel coordinates from the MOD14 file
plt.scatter(fp_longitude, fp_latitude, color='blue', label='FP Coordinates (MOD14)', s=1, alpha=0.5)

plt.title('Comparison of Geolocated Fire Pixels and FP Coordinates (MOD14)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()

# Show the plot
plt.show()
```
